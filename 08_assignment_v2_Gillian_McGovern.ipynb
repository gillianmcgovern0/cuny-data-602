Let's first modify the columns to make analyzing easier. There are percent signs and parentheses in the column names, so let's get rid of those and use underscores. Additionally, let's make all the column names lowercase.

```
# Create copy of the dataframe to make changes
df2 = df.copy()


# Renaming columns - use underscores and make lowercase
df2.rename(columns={'Category': 'category', 'Item': 'item', 'Serving Size': 'serving_size',
                   'Calories': 'calories', 'Calories from Fat': 'calories_from_fat', 'Total Fat': 'total_fat',
                   'Total Fat (% Daily Value)': 'total_fat_pct_daily_value', 'Saturated Fat': 'saturated_fat', 'Saturated Fat (% Daily Value)': 'saturated_fat_pct_daily_value'
                   , 'Trans Fat': 'trans_fat', 'Cholesterol': 'cholesterol', 'Cholesterol (% Daily Value)': 'cholesterol_pct_daily_value'
                   , 'Sodium': 'sodium', 'Sodium (% Daily Value)': 'sodium_pct_daily_value', 'Carbohydrates': 'carbohydrates'
                   , 'Carbohydrates (% Daily Value)': 'carbohydrates_pct_daily_value', 'Dietary Fiber': 'dietary_fiber'
                   , 'Dietary Fiber (% Daily Value)': 'dietary_fiber_pct_daily_value', 'Sugars': 'sugars', 'Protein': 'protein'
                   , 'Vitamin A (% Daily Value)': 'vitamin_a_pct_daily_value', 'Vitamin C (% Daily Value)': 'vitamin_c_pct_daily_value'
                   , 'Calcium (% Daily Value)': 'calcium_pct_daily_value', 'Iron (% Daily Value)': 'iron_pct_daily_value'}, inplace=True)
```

Before we start checking for improper values for the string columns, let's take a quick look at what values we're dealing with.

```
# Get a quick look at the different values for the string columns
print(df2["category"].unique())
print(df2["item"].unique())
```

This quick look showed us that it doesn't look like we have any encoded issues. We can still check for specials characters.

```
# Check for special characters

# Don't include `&` since we know that's included
print(df2[df2.category.str.contains(r'[@#$%+-/*]')])
# Don't include `&` since we know that's included
print(df2[df2.item.str.contains(r'[@#$%+-/*]')].item)
```

All of these values look valid, so let's keep the values as is.

Fix missing and invalid values in data:

We know that there are no issues with missing data, but `serving_size` contains values that are strings instead of numbers. We should update that via separating `serving_size` and converting number values to a numeric data type (as opposed to string).

```
# Split serving_size by " "
# This column has a messy format such that df3['serving_size'].str.split(" ") produces 5 different items
# Because of that, let's make the dataframe super wide, and then make some additional tranformations to make the dataframe tidy
# Let's make a copy of the df for now since we're going to make some transformations
df3 = df2.copy()
# Separate serving_size by " " so we'll have a column to represent the value, and the measuring unit
# That way, if we wanted, we can perform statistical analysis
df3[['serving_size', 'serving_size_unit', 'serving_size_other', 'serving_size_other_unit', 'other']] = df3['serving_size'].str.split(" ", expand=True).loc[:, 0:4]

# Remove the parenthesis from serving_size columns
df3['serving_size_other'] = df3['serving_size_other'].str.replace('(', '')
df3['serving_size_other_unit'] = df3['serving_size_other_unit'].str.replace(')', '')
df3['serving_size_other_unit'] = df3['serving_size_other_unit'].str.replace('(', '')

# Add back in "fl oz cup" wherever "fl" exists
df3["serving_size_unit"] = df3.serving_size_unit.str.replace("fl", "fl oz cup")
```

```
# Since the dataset sometimes contains 2 different serving size measurements, if the other measurement wasn't provided, let's just
# set the value to unavailable
# We won't use serving size for any data analysis so can probably just remove these "other" columns, but let's just set it to unavailable for now
df3.loc[df3.serving_size_other=="oz", "serving_size_other"] = np.nan
df3.loc[df3.serving_size_other_unit=="cup", "serving_size_other"] = np.nan
df3.loc[df3.serving_size_other_unit.isnull(), "serving_size_other"] = np.nan
print(df3[["serving_size_other", "serving_size_other_unit"]])
```

```
# Let's now clean up the last few cells
# For this row, the unit is in the generic "other" column
df3.loc[df3.serving_size_other_unit=="177", "serving_size_other_unit"] = df3.other
df3['serving_size_other_unit'] = df3['serving_size_other_unit'].str.replace(')', '')
df3['serving_size_other_unit'] = df3['serving_size_other_unit'].str.replace('(', '')

# Let's double check all the values
print(df3["serving_size_other"].unique())
print(df3["serving_size_other_unit"].unique())
print(df3["other"].unique())
```

```
# We now don't need `other` column so let's remove it
df3 = df3.drop(columns=["other"])
```

```
# Let's convert the serving size columns to numeric
df3["serving_size"] = pd.to_numeric(df3["serving_size"])
df3["serving_size_other"] = pd.to_numeric(df3["serving_size_other"])
```

```
# Let's look at the new summary of the data
df3.info()
```

```
from matplotlib import pyplot as plt
```

```
# Create a scatter plot of carbs vs calories

# Set colors based on category
colors = {'Breakfast': 'lightblue', 'Beef & Pork': 'coral', 'Chicken & Fish': 'black', 'Salads': 'limegreen', 'Snacks & Sides': 'red',
          'Desserts': 'orange', 'Beverages': 'yellow', 'Coffee & Tea': 'purple', 'Smoothies & Shakes': 'pink'}

# Plotting
# Change figure size
plt.figure(figsize=(10, 6))
# Call plt.scatter for each category using the corresponding color
for category, group in df3.groupby('category'):
    plt.scatter(group['carbohydrates'], group['calories'], color=colors[category], label=category)

# Add labels
plt.ylabel('Calories')
plt.xlabel('Carbohydrates')

# Add title
plt.title("Carbs vs Calories")

# Add legend and change position
plt.legend(loc='lower right')

# Display the plot
plt.show()
```

This scatter plot shows us the relationship between carbohydrates and calories. The general trend for this graph is that as carbohydrates increases, calories increases as well. Some of other findings from this graph:


*   Smoothies and Shakes (pink) has higher carbs and calories compared to other categories
*   Most items fall in the range 0 - 80 calories
*   Chicken and Fish has the highest amount of calories with an outlier at around 120 calories
*   Breakfast has a wide range of carbs with a few items having carbs in the 100-120 range


```
# Plot Protein vs Iron
# Update marker, color and size
plt.scatter(df3.protein, df3.iron_pct_daily_value, label="Protein vs Iron", marker="s", color="red", s=75)

# Add labels
plt.xlabel('Protein')
plt.ylabel('Iron (% Daily Value)')

# Add title
plt.title("Protein vs Iron")

# Display the plot
plt.show()
```

This scatter plot shows us the relationship between protein and iron. The general trend for this graph is that as protein increases, iron increase as well. So if a person needs more iron, they should pick an item with a high amount of protein.

```
# Plot the sodium amount for each category
plt.bar(df3.category, df3.sodium, label="Sodium")

# Add labels
plt.ylabel('Sodium')
plt.xlabel('Category')

# Add title
plt.title("Sodium by Category")

# Modify Axis Text Ticks/Labels
plt.xticks(rotation = 25)

# Display the plot
plt.show()
```

This bar plot shows us the sodium amount for each food category. Chicken and Fish contain the largest amount of sodium, followed by Breakfast, then Beef and Pork. What's also interesting is that Salads and Snacks and Sides share almost the same amount of sodium.

```
# Plot protein
# Colors for each attack type
plt.bar(df3.category, df3.protein, label="Protein", color="pink")

# Add labels and change fontsize
plt.xlabel('Category', fontsize=14)
plt.ylabel('Protein', fontsize=14)

# Add title
plt.title("Protein by Category")

# Modify Axis Text Ticks/Labels
plt.xticks(rotation = 25)

# Display the plot
plt.show()
```

Now let's create the same graphs via seaborn:

```
import seaborn as sns
```

```
# Create a scatter plot of carbs vs calories
# Update the color
colors = {'Breakfast': 'lightblue', 'Beef & Pork': 'coral', 'Chicken & Fish': 'black', 'Salads': 'limegreen', 'Snacks & Sides': 'red',
          'Desserts': 'orange', 'Beverages': 'yellow', 'Coffee & Tea': 'purple', 'Smoothies & Shakes': 'pink'}
plt.figure(figsize=(10, 6))
sns.scatterplot(x="carbohydrates", y="calories", data=df3, hue='category', palette=colors)

# Update labels (since sns already labels)
plt.ylabel('Calories')
plt.xlabel('Carbohydrates')

# Add title
plt.title("Carbs vs Calories")

# Add legend and change position
plt.legend(loc='lower right')

# Display the plot
plt.show()
```

```
# Plot protein vs iron
# Update marker, color, and point size
sns.scatterplot(x="protein", y="iron_pct_daily_value", data=df3, label="Protein vs Iron", marker="s", color="red", s=75)

# Add labels
plt.xlabel('Protein')
plt.ylabel('Iron (% Daily Value)')

# Add title
plt.title("Protein vs Iron")

# Display the plot
plt.show()
```

```
# Plot the sodium amount for each category
# To match the matplotlib graph, let's remove the error bar
sns.barplot(x="category", y="sodium", data=df3, color='blue', errorbar=None)

# Add labels
plt.ylabel('Sodium')
plt.xlabel('Category')

# Add title
plt.title("Sodium by Category")

# Modify Axis Text Ticks/Labels
plt.xticks(rotation = 25)

# Display the plot
plt.show()
```

```
# Plot the protein amount for each category
sns.barplot(x="category", y="protein", data=df3, color='pink', errorbar=None)

# Add labels
plt.ylabel('Protein')
plt.xlabel('Category')

# Add title
plt.title("Protein by Category")

# Modify Axis Text Ticks/Labels
plt.xticks(rotation = 25)

# Display the plot
plt.show()
```

*   To color the data points by category, need to call `plt.scatter` for each cateogry with the corresponding color (use for loop). For seaborn, you just need to call `sns.scatterplot` once.
